```python
import ipyplot
```

```python
ENV_LIST = ["BreakoutNoFrameskip-v4", "BankHeistNoFrameskip-v4", "WizardOfWorNoFrameskip-v4"]
NOISE_LIST = [0.0, 0.1, 0.2, 0.3, 0.4]
```

```python
for env_id in ENV_LIST:
    obs_gif_files = [f"./gifs/{env_id}--noise_{noise}--obs.gif" for noise in NOISE_LIST]
    reconstructed_gif_files = [f"./gifs/reconstructed/{env_id}--noise_{noise}--obs--reconstructed_v2.gif" for noise in NOISE_LIST]
    
    # Display gifs
    # gif_paths = [f"{gif_dir}/{exp_name}-run_{experiment}.gif" for experiment in EXPERIMENT_NOS]
    # labels = [f"run_{experiment}" for experiment in EXPERIMENT_NOS]
    gif_paths = obs_gif_files + reconstructed_gif_files
    ipyplot.plot_images(gif_paths, 
                        # labels, 
                        max_images=10, 
                        img_width=100, 
                        show_url=False)
```
