```python
from PIL import Image
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
orig_img = Image.open(Path('.') / 'dog.jpg')
```

```python
plt.figure(0)
plt.axis(False)
plt.imshow(orig_img)
```

```python
flipped_image = np.flipud(orig_img)
```

```python
plt.figure(1)
plt.axis(False)
plt.imshow(flipped_image)
```
